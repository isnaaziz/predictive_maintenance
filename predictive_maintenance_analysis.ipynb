# Proyek Mini: Prediksi Kegagalan Perangkat IoT (Predictive Maintenance)
### Mata Kuliah: Matematika Statistika

**Pipeline Predictive Maintenance – AI4I 2020 Dataset**

Proyek ini telah dimodularisasi dengan pipeline sebagai berikut:
1. **Data Loading & Validation**: Mengambil dan memvalidasi data
2. **EDA (Exploratory Data Analysis)**: Analisis data deskriptif dan visualisasi
3. **Statistical Modeling**: Pemodelan Regresi dan Distribusi Probabilitas
4. **Anomaly Detection**: Deteksi anomali menggunakan Z-score

Notebook ini mengintegrasikan modul-modul tersebut dalam struktur yang sesuai dengan pipeline di `main.py`.


```python
# Import Required Libraries & Modules
import os
import sys
from pathlib import Path
import time

# Pastikan src dapat diimport
sys.path.insert(0, str(Path.cwd()))

from src.data_loader import load_maintenance_data
from src.eda import perform_eda
from src.stats_models import analyze_regression, analyze_probability
from src.anomaly_detection import detect_anomalies

# Matplotlib inline untuk notebook
%matplotlib inline

# Konfigurasi output directory
OUTPUT_DIR = "output"
os.makedirs(OUTPUT_DIR, exist_ok=True)
print(f"✓ Output directory siap: {OUTPUT_DIR}/")
```

---

## Pipeline Execution

### STEP 1/4: Pemuatan & Validasi Data
Mengambil dan memvalidasi dataset AI4I 2020 dari UCI Repository.


```python
t_step = time.time()

# Load & validate data
df = load_maintenance_data(validate=True)

elapsed = time.time() - t_step
print(f"\n✓ Data loaded: {df.shape[0]} rows × {df.shape[1]} columns")
print(f"  Time: {elapsed:.2f}s\n")

# Display sample data
df.head()
```

### STEP 2/4: Exploratory Data Analysis (EDA)
Analisis statistik deskriptif dan visualisasi distribusi data kunci.


```python
t_step = time.time()

# Perform EDA with output directory
perform_eda(df, output_dir=OUTPUT_DIR)

elapsed = time.time() - t_step
print(f"\n✓ EDA selesai dalam {elapsed:.2f}s")
print(f"  Plots disimpan ke: {OUTPUT_DIR}/")
```

### STEP 3/4: Regresi & Korelasi
Menganalisis hubungan antara beban kerja (Torque) dan suhu proses menggunakan regresi linear.


```python
t_step = time.time()

# Analyze regression
analyze_regression(df, output_dir=OUTPUT_DIR)

elapsed = time.time() - t_step
print(f"\n✓ Analisis Regresi selesai dalam {elapsed:.2f}s")
```

### Distribusi Probabilitas (Normal & Poisson)
Menghitung peluang kejadian kritis dan kegagalan menggunakan distribusi probabilitas.


```python
t_step = time.time()

# Analyze probability distributions
analyze_probability(df, output_dir=OUTPUT_DIR)

elapsed = time.time() - t_step
print(f"\n✓ Analisis Probabilitas selesai dalam {elapsed:.2f}s")
```

### STEP 4/4: Deteksi Anomali (Z-Score)
Mendeteksi anomali pada sensor kecepatan rotasi menggunakan threshold Z-score.


```python
t_step = time.time()

# Detect anomalies
anomalies = detect_anomalies(df, output_dir=OUTPUT_DIR)

elapsed = time.time() - t_step
print(f"\n✓ Deteksi Anomali selesai dalam {elapsed:.2f}s")
print(f"  Total anomali terdeteksi: {anomalies.sum()}")

# Display anomaly results
print("\nSample data dengan anomali:")
anomalies.head(10)
```

Ringkasan hasil dari keseluruhan pipeline predictive maintenance.


```python
print("\n" + "╔" + "═"*58 + "╗")
print("║        PREDICTIVE MAINTENANCE PIPELINE COMPLETED         ║")
print("╚" + "═"*58 + "╝")
print(f"\n✓ Semua tahap pipeline selesai!")
print(f"  Output disimpan di: {OUTPUT_DIR}/")
print(f"\n📊 Dataset Info:")
print(f"  • Total records: {df.shape[0]:,}")
print(f"  • Total features: {df.shape[1]}")
print(f"\n🔍 Anomali Detection:")
print(f"  • Anomali terdeteksi: {anomalies.sum()} ({100*anomalies.mean():.2f}%)")
```
